```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
%matplotlib inline

import os
```

```python
base_dir = "/mnt/LIBOR/train/"
```

```python
for folder in os.listdir(base_dir):
    for i, filename in enumerate(os.listdir(os.path.join(base_dir, folder))):
        if i % 10 == 0:
            img = mpimg.imread(os.path.join(base_dir,folder,filename))
            plt.imshow(img)
            plt.title(folder)
            plt.show()
```
