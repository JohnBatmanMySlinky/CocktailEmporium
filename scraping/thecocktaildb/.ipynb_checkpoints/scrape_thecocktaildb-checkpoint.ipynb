```python
import wget
import urllib
import json
from collections import defaultdict
```

```python
# get json of glass types
with urllib.request.urlopen('https://www.thecocktaildb.com/api/json/v1/1/list.php?g=list') as url:
    glass_json = json.loads(url.read().decode())
    
# build dictionary of cocktails images associated with each glass type
glass_dict = defaultdict(list)
for glass_entry in glass_json['drinks']:
    # actual name of the glass type
    glass = glass_entry['strGlass'].lower()
    
    # get json with associated cocktails
    with urllib.request.urlopen('https://www.thecocktaildb.com/api/json/v1/1/filter.php?g='+glass.replace(' ','_')) as url:
        cocktail_json = json.loads(url.read().decode())
        
    # extract image url from json
    for cocktail in cocktail_json['drinks']:
        glass_dict[glass].append(cocktail["strDrinkThumb"])
```
